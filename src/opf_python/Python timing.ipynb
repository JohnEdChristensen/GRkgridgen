```python
import sc
import fcc
import bcc
import trig
import stet
import hx
import body_tet
import so
import base_ortho
import face_ortho
import body_ortho
import sm
import base_mono
import time
```

```python
sc_res = []
count = 0
start = time.time()
for n in range(1,100001):
    temp = sc.sc_srHNFs(n)
    count += len(temp)
    if n%1000==0:
        end = time.time()
        interval = end-start
        sc_res.append([n,interval,count])
        print("n: ",n," time: ",interval," count: ",count)
```

```python
fcc_res = []
count = 0
start = time.time()
for n in range(1,100001):
    temp = fcc.fcc_srHNFs(n)
    count += len(temp)
    if n%1000==0:
        end = time.time()
        interval = end-start
        fcc_res.append([n,interval,count])
        print("n: ",n," time: ",interval," count: ",count)
```

```python
fcc_res2 = []
count = 0
start = time.time()
for n in range(1,47):
    for m in [1,4,16]:
        if (n*n*n*m)<100000:
            temp = fcc.fcc_srHNFs(n*n*n*m)
            count += len(temp)
            end = time.time()
            interval = end-start
            fcc_res2.append([n*n*n*m,interval,count])
            print("n: ",n*n*n*m," time: ",interval," count: ",count)
```

```python
fo_res = []
count = 0
start = time.time()
for n in range(1,100001):
    temp = face_ortho.face_ortho_srHNFs(n)
    count += len(temp)
    if n%1000==0:
        end = time.time()
        interval = end-start
        fo_res.append([n,interval,count])
        print("n: ",n," time: ",interval," count: ",count)
```

```python
import matplotlib.pyplot as plt
```

```python
xf = []
tf = []
nf = []
for f in fo_res:
    xf.append(f[0])
    tf.append(f[1])
    nf.append(f[2])
    
xf1 = []
tf1 = []
nf1 = []
for f in fcc_res:
    xf1.append(f[0])
    tf1.append(f[1])
    nf1.append(f[2])
    
xf2 = []
tf2 = []
nf2 = []
for f in fcc_res2:
    xf2.append(f[0])
    tf2.append(f[1])
    nf2.append(f[2])
```

```python
plt.plot(xf,tf,label="fco")
plt.plot(xf1,tf1,label="fcc all n")
plt.plot(xf2,tf2,label="fcc relevant n")
plt.xscale('log')
plt.yscale('log')
plt.title("Comulative Time scale for generating srHNFs")
plt.ylabel("Time (sec)")
plt.xlabel("Volume factor (n)")
plt.legend()
plt.savefig("srHNF_algorithm_timing.pdf")
plt.show()
```

```python
plt.plot(xf,nf,label="fco")
plt.plot(xf1,nf1,label="fcc all n")
plt.plot(xf2,nf2,label="fcc relevant n")
plt.xscale('log')
plt.yscale('log')
plt.title("Comulative number ofsrHNFs")
plt.ylabel("Time (sec)")
plt.xlabel("Volume factor (n)")
plt.legend()
plt.savefig("srHNFs_number.pdf")
plt.show()
```
